```python
def pipeline(df):
   #Seting up a cast dictionary for mapping the columns on scale 0 - 5
    cast_1 = {
        'Ex': 5,
        'TA': 3,
        'Gd': 4,
        'Fa': 2,
        'Po': 1,
           0: 0
    }
    
    cast_2 = {
        'Gd': 4,
        'Av': 3,
        'Mn': 2,
        'No': 1,
           0: 0
    }
    
    #Grabbing the columns for transformation
    some_features = ['lot_frontage', 'lot_area', 'street', 'utilities', 'condition_1',
       'bldg_type', 'overall_qual', 'overall_cond', 'year_built',
       'year_remod/add', 'exter_qual', 'exter_cond',
       'bsmt_qual', 'bsmt_cond', 'bsmt_exposure', 'total_bsmt_sf', 'heating',
       'heating_qc', '1st_flr_sf', '2nd_flr_sf', 'full_bath', 'half_bath',
       'bedroom_abvgr', 'kitchen_abvgr', 'kitchen_qual', 'totrms_abvgrd',
       'garage_yr_blt', 'garage_area', 'garage_qual', 'garage_cond',
       'paved_drive', 'pool_area', 'pool_qc']
    
    X = df[some_features]
    
    #filling missing values with '0'
    X.fillna(0, inplace = True)
    
    #columns to binarize:
    X['street'] = X['street'].map({'Pave': 1, 'Grvl': 0})
    X['utilities'] = X['utilities'].map({'AllPub': 1, 'NoSeWa': 0 ,
                                              'NoSewr': 0, 'ELO':0})
    X['heating'] = X['heating'].map({'GasA': 1, 'Floor': 0, 
                                     'GasW': 0 ,'Grav': 0, 
                                     'OthW': 0, 'Wall': 0})
    X['pool_qc'] = X['pool_qc'].map({'Ex': 1, 'Gd': 1, 'TA': 1, 'Fa': 1, 0: 0})

    #extrapolating age from dataset, base max year built houses (2010):
    X['year_built']     = 2010 - X['year_built']
    X['year_remod/add'] = 2010 - X['year_remod/add']
    
    #mapping the cast dict to features scale 0 - 5:
    X['garage_qual']     = X['garage_qual'].map(cast_1)
    X['garage_cond']     = X['garage_cond'].map(cast_1)
    X['bsmt_qual']      = X['bsmt_qual'].map(cast_1)
    X['bsmt_cond']      = X['bsmt_cond'].map(cast_1)
    X['exter_qual']     = X['exter_qual'].map(cast_1)
    X['exter_cond']     = X['exter_cond'].map(cast_1)
    X['heating_qc']     = X['heating_qc'].map(cast_1)
    X['kitchen_qual']   = X['kitchen_qual'].map(cast_1)
    X['bsmt_exposure']   = X['bsmt_exposure'].map(cast_2)

    #getting dummies:
    X = pd.get_dummies(X, columns = ['condition_1',
                                     'bldg_type', 
                                     'paved_drive'], drop_first = True)

    #any feature interactions go here: 
    X['total_sf'] = X['total_bsmt_sf'] + X['1st_flr_sf'] + X['2nd_flr_sf']
    X['overall_score'] = X['overall_qual'] * X['overall_cond']
    X['exter_score'] = X['exter_qual'] * X['exter_cond']
    X['bsmt_score'] = X['bsmt_qual'] * X['bsmt_cond']
    X['garage_score'] = X['garage_qual'] * X['garage_cond']

    return X
```

**Modeling Time! Let's Work It!**

```python
features = ['lot_frontage', 'lot_area', 'street', 'utilities', 'overall_qual',
       'overall_cond', 'year_built', 'year_remod/add', 'exter_qual',
       'exter_cond', 'bsmt_qual', 'bsmt_cond', 'bsmt_exposure',
       'total_bsmt_sf', 'heating', 'heating_qc', '1st_flr_sf', '2nd_flr_sf',
       'full_bath', 'half_bath', 'bedroom_abvgr', 'kitchen_abvgr',
       'kitchen_qual', 'totrms_abvgrd', 'garage_yr_blt', 'garage_area',
       'garage_qual', 'garage_cond', 'pool_area', 'pool_qc',
       'condition_1_Feedr', 'condition_1_Norm', 'condition_1_PosA',
       'condition_1_PosN', 'condition_1_RRAe', 'condition_1_RRAn',
       'condition_1_RRNe', 'condition_1_RRNn', 'bldg_type_2fmCon',
       'bldg_type_Duplex', 'bldg_type_Twnhs', 'bldg_type_TwnhsE',
       'paved_drive_P', 'paved_drive_Y', 'total_sf', 'overall_score',
       'exter_score', 'bsmt_score', 'garage_score']

X = df_some_feat[features]
y = df_some_feat['saleprice']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#As opposed to last time with 'least_features', this time I will scale my data:

sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)

lr_1 = LinearRegression()
lr_1.fit(Z_train, y_train)
lr_1.score(Z_train, y_train), lr_1.score(Z_test, y_test)
```

```python
lr_1.score(Z_test, y_test) - lr_1.score(Z_train, y_train)
```

```python
#Lasso:
lasso = LassoCV()
lasso.fit(Z_train, y_train)
lasso.score(Z_train, y_train), lasso.score(Z_test, y_test)
```

```python
lasso.score(Z_test, y_test) - lasso.score(Z_train, y_train)
```

```python
#Ridge:
ridge = RidgeCV(alphas = np.linspace(.1, 10, 100))
ridge.fit(Z_train, y_train)
ridge.score(Z_train, y_train), ridge.score(Z_test, y_test)
```

```python
ridge.score(Z_test, y_test) - ridge.score(Z_train, y_train)
```
