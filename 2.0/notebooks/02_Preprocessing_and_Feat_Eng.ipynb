**Reading in Dummified Data**

## Addressing Remaining Null Values

```python
ames_df.isna().sum().sort_values(ascending = False)[0:15]
```

```python
ames_test_df.isna().sum().sort_values(ascending = False)
```

```python
#checking nulls for 'lot_frontage'
ames_df[ames_df['lot_frontage'].isna()].head()
```

- If Lot Frontage is null for a house one would assume the property is not adjacent to a road, however many of the homes above have garages. I am therefore going to replace the 'lot_frontage' nulls with the median value. 

```python
ames_df['lot_frontage'].median(), ames_test_df['lot_frontage'].median()
```

```python
ames_df['lot_frontage'].fillna(value = ames_df['lot_frontage'].median(), inplace = True)
```

```python
#repeating for the testing data:
ames_test_df['lot_frontage'].fillna(value = ames_test_df['lot_frontage'].median(), inplace = True)
```

```python
#checking nulls for 'garage_yr_blt'
ames_df[ames_df['garage_yr_blt'].isna()][['garage_yr_blt', 'garage_cars', 'garage_area']].head()
```

```python
ames_df[ames_df['garage_yr_blt'].isna()][['garage_yr_blt', 'garage_cars', 'garage_area']].mean()
```

- If Garage Year Built is null for a house one would assume that there is no garage. I can't .fillna() with a value of 0, it doesn't make sense for what the value represents. As with 'lot_frontage', I am going to replace the 'garage_yr_blt' nulls with the median value. 

```python
ames_df['garage_yr_blt'].median(), ames_test_df['garage_yr_blt'].median()
```

```python
ames_df['garage_yr_blt'].fillna(value = ames_df['garage_yr_blt'].median(), inplace = True)
```

```python
#repeating for the testing data:
ames_test_df['garage_yr_blt'].fillna(value = ames_test_df['garage_yr_blt'].median(), inplace = True)
```

```python
#checking nulls for 'msn_vnr_area'
ames_df[ames_df['mas_vnr_area'].isna()][['mas_vnr_type_BrkCmn', 'mas_vnr_type_BrkFace', 'mas_vnr_type_None',
                                         'mas_vnr_type_Stone', 'mas_vnr_type_nan']].head()
```

```python
ames_df[ames_df['mas_vnr_area'].isna()][['mas_vnr_type_BrkCmn', 'mas_vnr_type_BrkFace', 'mas_vnr_type_None',
                                         'mas_vnr_type_Stone', 'mas_vnr_type_nan']].mean()
```

- If Masonry Veneer Area is null for a house one would assume that there is no exterior masonry. This is further confirmed by the fact that 'mas_vnr_type' is null for all instances where 'mas_vnr_area' is null. These null values can be filled with zero.

```python
ames_df['mas_vnr_area'].fillna(0, inplace = True)
```

```python
#repeating for the testing data:
ames_test_df['mas_vnr_area'].fillna(0, inplace = True)
```

- At this point all the null values in the testing data have been address, however there still remain a few null values in the training data which need to be taken care of. 

```python
ames_df.isna().sum().sort_values(ascending = False)[0:10]
```

```python
#checking nulls for 'bsmt_full_bath'
ames_df[ames_df['bsmt_full_bath'].isna()]
```

- Both of the above have null or zero values for the features/columns related to a House's basement. I can impute or .fillna() these values to be zero. 

```python
ames_df['bsmt_full_bath'].fillna(value = 0, inplace = True)
ames_df['bsmt_half_bath'].fillna(value = 0, inplace = True)
ames_df['bsmtfin_sf_1'].fillna(value = 0, inplace = True)
ames_df['bsmtfin_sf_2'].fillna(value = 0, inplace = True)
ames_df['bsmt_unf_sf'].fillna(value = 0, inplace = True)
ames_df['total_bsmt_sf'].fillna(value = 0, inplace = True)
```

```python
#checking nulls for 'garage_area'
ames_df[ames_df['garage_area'].isna()]
```

- If Garage Area or Garage Cars is null for a house one would assume that there is no garage but the data here indicates there is a garage, as a 'garage_yr_built' is recorded. I am going to impute the values for 'garage_area' and 'garage_car' to be the median value.

```python
ames_df['garage_area'].median(), ames_df['garage_cars'].median()
```

```python
ames_df['garage_area'].fillna(ames_df['garage_area'].median(), inplace = True)
ames_df['garage_cars'].fillna(ames_df['garage_area'].median(), inplace = True)
```
